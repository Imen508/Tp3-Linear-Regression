```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

<h3><span style="color:blue"> Example 1 </span></h3>

```python
# Data loading
df = pd.read_csv('grav.csv')
df.head()
```

```python
# Selection of the first column of our dataset (height)
X = df.iloc[0:len(df),0]
# Selection of second columns of our dataset (seconds squared)
Y = df.iloc[0:len(df),1] 
```

```python
axes = plt.axes()
# Draw a grid
axes.grid()  
# X and Y are the variables that we extracted in the previous part
plt.scatter(X,Y) 
plt.show()
```

```python
from scipy import stats

# Linregress () returns multiple return variables - We are particularly interested in slope and intercept
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)
def predict(x):
    return slope * x + intercept
# The fitLine variable will be an array of predicted values from the X array
fitLine = predict(X)
plt.plot(X, fitLine, c='r')
```

```python
# Print the value of 'slope'
print (slope)
```

```python
# Print the value of 'intercept'
print (intercept)
```

```python
# Data loading
import pyreadstat

df, metadata = pyreadstat.read_sav("test_scores.sav")
df.head(10)
```

```python
# Shape of our dataset
df.shape
# Info our dataset
df.info()
# Describe our dataset
df.describe()
```

```python
# Using pairplot we'll visualize the data for correlation
sns.pairplot(df, x_vars=['school','school_setting','school_type','classroom','teaching_method','n_student','student_id','gender', 'lunch','pretest'], 
             y_vars='posttest', size=4, aspect=1, kind='scatter')
plt.show()
```

```python
# Visualizing the data using heatmap
sns.heatmap(df.corr(), cmap="YlGnBu", annot = True)
plt.show()
```

```python
# Creating X and y
X = df['pretest']
y = df['posttest']
plt.scatter(X,y)
plt.xlabel("pretest")
plt.ylabel("posttest")
```

```python
# Splitting the varaibles as training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

```python
# Take a look at the train dataset
X_train
y_train
```

```python
# Importing Statsmodels.api library from Stamodel package
import statsmodels.api as sm

# Adding a constant to get an intercept
X_train_sm = sm.add_constant(X_train)
```

```python
# Fitting the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()
# Printing the parameters
lr.params
```

```python
# Performing a summary to list out all the different parameters of the regression line fitted
lr.summary()
```

```python
# Visualizing the regression line
plt.scatter(X_train, y_train)
plt.plot(X_train, 13.3750 + 0.9783*X_train, 'r')
plt.show()
```

```python
# Predicting y_value using traingn data of X
y_train_pred = lr.predict(X_train_sm)
# Creating residuals from the y_train data and predicted y_data
res = (y_train - y_train_pred)
```

```python
# Plotting the histogram using the residual values
fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```

```python
# Looking for any patterns in the residuals
plt.scatter(X_train,res)
plt.show()
```

```python
# Adding a constant to X_test
X_test_sm = sm.add_constant(X_test)
# Predicting the y values corresponding to X_test_sm
y_test_pred = lr.predict(X_test_sm)
# Printing the first 15 predicted values
y_test_pred
```

```python
# Importing r2_square
from sklearn.metrics import r2_score

# Checking the R-squared value
r_squared = r2_score(y_test, y_test_pred)
r_squared
```

```python
# Visualize the line on the test set
plt.scatter(X_test, y_test)
plt.plot(X_test, y_test_pred, 'r')
plt.show()
```

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split

X_train_lm, X_test_lm, y_train_lm, y_test_lm = train_test_split(X, y, train_size = 0.7, 
                                                                test_size = 0.3, random_state = 100)
```

```python
# Shape of the train set without adding column
X_train_lm.shape
# Adding additional column to the train and test data
X_train_lm = X_train_lm.values.reshape(-1,1)
X_test_lm = X_test_lm.values.reshape(-1,1)
print(X_train_lm.shape)
print(X_test_lm.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Creating an object of Linear Regression
lm = LinearRegression()
# Fit the model using .fit() method
lm.fit(X_train_lm, y_train_lm)
```

```python
# Intercept value
print("Intercept :",lm.intercept_)
# Slope value
print('Slope :',lm.coef_)
```

```python
# Making Predictions of y_value
y_train_pred = lm.predict(X_train_lm)
y_test_pred = lm.predict(X_test_lm)
# Comparing the r2 value of both train and test data
print(r2_score(y_train,y_train_pred))
print(r2_score(y_test,y_test_pred))
```

```python
# Loading the data
boston = pd.read_csv('housing.data.txt' , sep=",")
print("housing's shape : ", boston.shape)
# Print the first 10 rows of the data
boston.head(10)
```

<h3><span style="color:orange"> Data preprocessing </span></h3>

```python
boston.isnull().sum()
```

<h3><span style="color:orange"> Data Analysis </span></h3>

L'analyse exploratoire des donnÃ©es est une Ã©tape trÃ¨s importante avant la formation du modÃ¨le. Dans cette section, nous utiliserons quelques visualisations pour comprendre la relation de la variable cible avec d'autres fonctionnalitÃ©s.

```python
# Plot the distribution of the target variable MEDV
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(boston['MEDV'], bins=30)
plt.show()
```

<h4><span style="color:Purple"> Remarque: </span></h4>

On voit que les valeurs de MEDV sont distribuÃ©es normalement avec peu de valeurs aberrantes.

> Ensuite, nous crÃ©ons une matrice de corrÃ©lation qui mesure les relations linÃ©aires entre les variables. La matrice de corrÃ©lation peut Ãªtre formÃ©e en utilisant la fonction de la bibliothÃ¨que pandas dataframe. Nous utiliserons la fonction heatmap de la bibliothÃ¨que seaborn pour tracer la matrice de corrÃ©lation.

```python
correlation_matrix = boston.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True)
```

> Le coefficient de corrÃ©lation va de -1 Ã  1. Si la valeur est proche de 1, cela signifie qu'il existe une forte corrÃ©lation positive entre les deux variables. Lorsqu'il est proche de -1, les variables ont une forte corrÃ©lation nÃ©gative.

<h4><span style="color:Purple"> Remarque: </span></h4>

 Pour ajuster un modÃ¨le de rÃ©gression linÃ©aire, nous sÃ©lectionnons les entitÃ©s qui ont une forte corrÃ©lation avec notre variable cible MEDV. En regardant la matrice de corrÃ©lation, nous pouvons voir que RM a une forte corrÃ©lation positive avec MEDV (0,7) alors que LSTAT a une forte corrÃ©lation nÃ©gative avec MEDV (-0,74).

 Un point important dans la sÃ©lection des entitÃ©s pour un modÃ¨le de rÃ©gression linÃ©aire est de vÃ©rifier la multi-colinÃ©aritÃ©. Les fonctionnalitÃ©s RAD, TAX ont une corrÃ©lation de 0,91. Ces paires de caractÃ©ristiques sont fortement corrÃ©lÃ©es les unes aux autres. Nous ne devons pas sÃ©lectionner ces deux fonctionnalitÃ©s ensemble pour entraÃ®ner le modÃ¨le. Il en va de mÃªme pour les fonctionnalitÃ©s DIS et AGE qui ont une corrÃ©lation de -0,75.

```python
plt.figure(figsize=(20, 5))

features = ['LSTAT', 'RM']
target = boston['MEDV']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = boston[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

<h4><span style="color:Purple"> Remarque: </span></h4>

Les prix augmentent Ã  mesure que la valeur de RM augmente linÃ©airement. Il y a peu de valeurs aberrantes et les donnÃ©es semblent plafonnÃ©es Ã  50.
Les prix ont tendance Ã  baisser avec une augmentation de LSTAT. Bien qu'il ne semble pas suivre exactement une ligne linÃ©aire.

<h3> <span style="color:orange">Preparing the data for training the model</span> </h3>

> Nous concatÃ©nons les colonnes LSTAT et RM en utilisant np.c_provided par la bibliothÃ¨que numpy.

```python
X = pd.DataFrame(np.c_[boston['LSTAT'], boston['RM']], columns = ['LSTAT','RM'])
Y = boston['MEDV']
```

<h3><span style="color:orange"> Splitting the data into training and testing sets </span></h3>

Ensuite, nous divisons les donnÃ©es en ensembles d'entraÃ®nement et de test. Nous formons le modÃ¨le avec 70% des Ã©chantillons et testons avec les 30% restants. Nous faisons cela pour Ã©valuer les performances du modÃ¨le sur des donnÃ©es invisibles. Pour diviser les donnÃ©es, nous utilisons la fonction train_test_split fournie par la bibliothÃ¨que scikit-learn. Nous imprimons enfin les tailles de notre ensemble d'entraÃ®nement et de test pour vÃ©rifier si le fractionnement s'est produit correctement.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

<h3><span style="color:orange"> Training and testing the model </span></h3>

Nous utilisons LinearRegression de scikit-learn pour entraÃ®ner notre modÃ¨le sur les ensembles de formation et de test.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

<h3><span style="color:orange"> Model evaluation</span> </h3>

Nous Ã©valuerons notre modÃ¨le en utilisant RMSE et R2-score.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Model evaluation for training set
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# Model evaluation for testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
